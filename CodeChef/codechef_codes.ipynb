```python
#lapindromes
t = int(input())

for _ in range(t):
    s = input()

    n = len(s)

    if n % 2 == 0:
        left = s[:n//2]
        right = s[n//2:]
    else:
        left = s[:n//2]
        right = s[n//2+1:]

    if sorted(left) == sorted(right):
        print("YES")
    else:
        print("NO")
```

```python
#Reverse The Number
num = int(input("Enter a number: "))

reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

print("Reversed number:", reverse)
```

```python
#turbo sort784
n = int(input())

arr = []

for i in range(n):
    arr.append(int(input()))

arr.sort()

for num in arr:
    print(num)
```
